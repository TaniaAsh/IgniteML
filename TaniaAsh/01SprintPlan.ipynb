### План работы на первый спринт (11-24/01/25)
**Модель:** Линейная Регрессия  
**Датасет:** [House Price Data](https://www.kaggle.com/datasets/jakewright/house-price-data)

1. **Присоединить данные**  
   Загрузить данные из источника, организовать их в DataFrame.

2. **Изучить датасет**  
   - Определить назначение колонок.  
   - Оценить разброс данных, выделить аутлайнеры.  
   - Провести визуализацию зависимостей между ключевыми переменными.

3. **Почистить данные**  
   - Убрать лишние строки.  
   - Выделить и обработать аутлайнеры и неверные значения.  
   - Заменить категориальные данные числовыми, если необходимо.  

4. **Создать базовую модель**  
   - Выбрать входные параметры (X) и целевой параметр (y).  
   - Построить модель линейной регрессии на основе X и y.  
   - Отобразить результаты на графике.

5. **Проверка модели**  
   - Провести тестирование на отдельных примерах из датасета.  

6. **Расширить модель**  
   - Построить модель с использованием нескольких фичей.  

7. **Работа с репозиторием**  
   - Выложить проект на GitHub:  
     - В личный репозиторий.  
     - В репозиторий проекта (сохранить в своей папке).

8. **Отчет о работе**  
   Написать Personal Reflection (2-4 абзаца) о выполненных задачах и результатах. **Желательно на английском.**
